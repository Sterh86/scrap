## Matplotlib

Очен много примеров на: https://matplotlib.org/stable/gallery/index.html

````python
import matplotlib.pyplot as plt
````

````python
 # показывает все графики в нотебуке
````

````python
%matplotlib inline
````

````python
import numpy as np
````

````python
x = np.linspace(0,6)
````

````python
y = x**2
````

````python
plt.plot(x, y, 'b') # 'b' синий цвет
plt.xlabel('Имя абсциссы ')
plt.ylabel('Имя ординаты')
plt.title('Имя графика')
plt.show()
````

````python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') 
plt.subplot(1,2,2)
plt.plot(y, x, 'g*')
````

**Более мощный метод**

Позволяет лучше контролировать график

````python
# Создать белое пустое поле
fig = plt.figure()

# Добавить оси (позиция начала х, позиция начала у, длина х, бысота у) все позиции от 0 до 1
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Нарисовать на этих осях
axes.plot(x, y, 'b')
axes.set_xlabel('Имя абсциссы') # здесь используется глагол set (установить)
axes.set_ylabel('Имя ординаты')
axes.set_title('Имя графика');
````

````python
# Более автоматический вариант
fig, axes = plt.subplots()

# Исползуем оси для добабления информации
axes.plot(x, y, 'b')
axes.set_xlabel('Имя абсциссы')
axes.set_ylabel('Имя ординаты')
axes.set_title('Имя графика');
````

````python
fig, axes = plt.subplots(figsize=(12,5)) # изменяем размер графика

axes.plot(x, y, 'b')
axes.set_xlabel('Имя абсциссы')
axes.set_ylabel('Имя ординаты')
axes.set_title('Имя графика');
````

````python
z = y-5
fig, axes = plt.subplots(figsize=(12,5)) 

axes.plot(x, y, 'b',label='График 1')
axes.plot(x,z,'r',label='График 2')
axes.set_xlabel('Имя абсциссы')
axes.set_ylabel('Имя ординаты')
axes.set_title('Имя графика')
axes.legend(loc=0);
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(x, y, x, z)
axes[0].set_title("Автоматический диапазон координат")


axes[1].plot(x, y, x, z)
axes[1].set_ylim([0, 20])
axes[1].set_xlim([1, 4])
axes[1].set_title("Мануалный диапазон координат");
````

````python
plt.scatter(x,y); # Собрание пунктов
````

## Seaborne

Визуализация со специалным упором на статистику. Галерея: https://seaborn.pydata.org/examples/index.html

````python
import seaborn as sns
````

````python
data = np.random.randn(100)
````

````python
sns.histplot(data) 
````
